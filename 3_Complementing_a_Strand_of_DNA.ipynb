Nesse exercício, teremos que obter o reverso da fita complementar do DNA.

Umas das características do DNA é a sua **complementariedade de bases**. Se em uma fita tiver uma adenina, na outra fita encontraremos uma timina (**A-T**); se for uma guanina, na outra fita teremos uma citosina (**G-C**), e vice-versa.

Dessa maneira, uma **base púrica** (A, G), que tem 2 aneis, sempre irá se ligar a uma **base pirimídica** (T, C), que tem um anel. Essa ligação entre bases é feita por ligações de hidrogênio. Se você já se questionou qual o sentido da vida, os biólogos talvez tenham uma resposta: é **5' - 3'** *(5 linha, 3 linha)*.

Isso é devido ao sentido da síntese dos ácidos nucleicos (DNA e RNA) que acontece da extremidade 5' para a extremidade 3'. Mas que sentido é esse?


image.png

O nucleotídeo é divido em 3 partes: um **grupo fosfato**, uma **pentose** (açúcar de 5 carbonos) e uma **base nitrogenada**. O sentido 5'-3' está relacionado aos carbonos da pentose nas posições 5' e 3'. Repare na figura abaixo que o **carbono 3' possui um radical hidroxila (OH)**, enquanto que o **carbono 5' está ligado ao grupo fosfato**.



image.png

Radicais OH possuem um desequilíbrio de elétrons e isso os tornam altamente reativos (são agentes oxidantes), os seja, estão "doidos" para se ligarem a alguma outra molécula. E é justamente nesse carbono 3' que o grupo fosfato do próximo nucleotídeo irá se ligar, por meio de uma **ligação fosfodiester**, promovendo o crescimento da fita de DNA (ou RNA). Portanto, para que a fita de DNA cresça, a enzima DNA polimerase (que é quem adiciona os nucleotídeos) **só consegue adicionar o próximo nucleotídeo no carbono 3'**. 

Dessa maneira, em uma extremidade do DNA, temos o grupo fosfato ligado ao carbono 5', e na outra extremidade, temos um OH ligado ao carbono 3'.

##**Problema:**

http://rosalind.info/problems/revc/

Given: A DNA string s of length at most 1000 bp.

Return: The reverse complement sc of s.

Sample Dataset:

```
# string s fornecida pelo exemplo
s = "AAAACCCGGT"
```

Sample Output:

```
# criar um dicionário com o complementar de cada base
troca= {"A": "T",
       "T": "A",
       "G": "C",
       "C": "G"}
```

```
# usei list comprehension para iterar sobre cada nucleotídeo de "s" e fazer a troca utilizando o dicionário "troca"
s_complementar = [troca[nt] for nt in s]
print(s_complementar)
```

```
# agora temos que concatenar os elementos da lista e revertê-la, usando "join()" e "reversed()"
s_complementar = "".join(reversed(s_complementar))
print(s_complementar)
```
