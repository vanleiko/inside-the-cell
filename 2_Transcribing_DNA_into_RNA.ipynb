#**Transcrevendo o DNA em RNA**
___

Nesse exercício, iremos fazer a transcrição do DNA, ou seja, substituir as timinas por uracilas, e assim obter um transcrito de RNA.

Assim como o DNA, o **RNA** (ácido ribonucleico) também é constitutído por nucleotídeos. Porém, há algumas diferenças em sua constituição em relação ao DNA:

- o RNA também possui um **grupo fosfato**, que confere o caráter ácido à molécula;

- a **pentose** do RNA é a **ribose**: uma pentose que possui o grupo hidroxila (OH) no carbono 2';

- as **bases nitrogenadas** do RNA poderão ser a **adenina**,  **guanina** *(bases púricas: 2 aneis)*, **citosina** e **uracila** *(bases pirimídicas: 1 anel)*.

- Além disso, o **RNA** é uma **fita simples** *(na maioria das vezes, pois há virus que possuem genoma de RNA fita dupla)*, enquanto que o DNA é uma fita dupla *(na maioria das vezes, pois há virus que possuem genoma de DNA fita simples)*.


O RNA é constituído por dezenas, centenas ou milhares de nucleotídeos, que podem ser de **quatro tipos**, e estes recebem o nome de acordo com a base nitrogenada que possuírem: **A**denina, **G**uanina, **C**itosina ou **U**racila. Quem sintetiza o RNA no processo da transcrição é a enzima RNA polimerase.

Há vários tipos de RNAs. Mas podemos dividí-los em 2 grandes grupos: **RNAs codificantes**, que serão traduzidos em proteínas, e **RNAs não-codificantes (ncRNA)**, que não serão traduzidos em proteínas. 

Os **RNAs codificantes** são os RNAs mensageiros (mRNA), que irão servir de molde para a síntese de proteínas, ou seja, eles carregam a informação, a receita que estava armazenada lá no DNA de como será a sequência de aminoácidos da proteína. Já os **RNAs não-codificantes** irão atuar em vários processos, como no transporte de aminoácidos para a síntese proteica (RNA transportador, tRNA), na constituição dos ribossomos, que são as fábricas de proteínas (RNA ribossômico, rRNA), e também há aqueles RNAs que atuarão na regulação da expressão de genes (miRNA, siRNA, snRNA, etc).

*Mas por que fazemos a transcrição do DNA para RNA? Por que não sintetizamos proteínas direto da fita de DNA?*

Imagine que você tem milhões de receitas salvas no seu computador e um amigo pede a você a receita de bolo de chocolate. O que você faz? Você leva o seu HD até a casa dele, que armazena suas milhões de receitas, correndo o risco de perder esse HD ou danificá-lo? Ou você simplesmente faz uma cópia apenas da receita do bolo de chocolate e envia a ele por meio de um pendrive, não importando se ele perder esse pendrive, pois você tem a receita original salva? Esse é um dos motivos por que a célula faz a transcrição da informação presente no DNA para uma molécula intermediária, o RNA. 

Imagine se a célula tivesse que ficar abrindo a fita de DNA toda vez que precisasse de uma proteína. Isso deixaria a fita de DNA exposta por muito tempo e vulnerável a mutações ou degradação. Por isso, é mais seguro transferir essa informação para uma molécula intermediária, o mRNA, que poderá ser lida quantas vezes for necessário para a síntese de proteínas e, posteriormente, esse mRNA será degradado quando a célula já tiver a quantidade adequada de proteínas.

Uma outra observação: você reparou que no nosso exercício, foi pedido que apenas trocássemos o "T" pelo "U", sem alterar os demais nucleotídeos? Mas se a transcrição tem como molde o DNA, não deveríamos ter feito o RNA complementar ao trecho de DNA transcrito? Ou seja, se tínhamos no DNA "GATGGAACTTGACTACGTAAATT", o RNA resultante não deveria ser "CUACCUUGAACUCAUUUAA"? Porém, nosso resultado foi "GAUGGAACUUGACUACGUAAAUU". Isso por que, como mostra a figura abaixo, no DNA nós temos a **fita molde** *(template strand)* e a **fita codificadora** *(coding strand)*. A fita molde servirá de fato como o molde para a síntese do RNA, mas repare que a sequência do RNA recém obtida é IGUAL à sequência da fita codificadora do DNA, diferindo apenas na timina/uracila! Por isso, por convenção, chamamos essa fita de fita codificadora. Portanto, o trecho de DNA fornecido pelo exercício foi da **fita codificadora!**.
