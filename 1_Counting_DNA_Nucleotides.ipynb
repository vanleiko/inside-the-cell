#**Contando nucleotídeos do DNA**
___

Nesse exercício, iremos contar a quantidade de cada tipo de nucleotídeo presente no DNA.

Há dois tipos de **ácidos nucleicos** nos seres vivos: o **DNA** (ácido desoxirribonucleico) e o **RNA** (ácido ribonucleico).

O nome "ácido" se deve porque essas moléculas possuem caráter ácido, e "nucleico" porque foram primeiramente descobertas no núcleo da célula (no início, foram chamadas de "nucleínas").

**>>> Curiosidade: foi no século XIX que o bioquímico Johann Miescher descobriu que o núcleo das células possuía várias moléculas ácidas e ricas em fósforo e nitrogênio, denominando-as de "nucleínas". Miescher estudou os glóbulos brancos, pois essas células possuem grandes núcleos, fáceis de serem isolados e estudados. E sabe como Miescher conseguia esses glóbulos brancos? Indo até hospitais e coletando ataduras cheias de pus dos pacientes.**

A unidade fundamental do DNA é o **nucleotídeo**, que pode ser dividido em 3 partes:

- um **grupo fosfato**, que confere o caráter ácido à molécula de DNA;

- uma **pentose**, que é um açúcar de 5 carbonos (no caso do DNA, é a desoxirribose: uma pentose que perdeu o oxigênio do carbono 2', *lê-se "carbono 2 linha"*);

- uma **base nitrogenada**, que para o DNA, pode ser **adenina**,  **guanina** *(bases púricas: 2 aneis)*, **citosina** e **timina** *(bases pirimídicas: 1 anel)*.

Portanto, o DNA pode ser constituído por milhares, milhões, e até bilhões, de nucleotídeos, que podem ser de **quatro tipos**, e estes recebem o nome de acordo com a base nitrogenada que possuírem: **A**denina, **T**imina, **G**uanina ou **C**itosina.

Given: A DNA string s of length at most 1000 nt.

Return: Four integers (separated by spaces) counting the respective number of times that the symbols 'A', 'C', 'G', and 'T' occur in s.

Sample dataset:

```
# string s fornecida como exemplo
s = "AGCTTTTCATTCTGACTGCAACGGGCAATATGTCTCTGTGTGGATTAAAAAAAGAGTGTCTGATAGCAGC"
```

Sample output:

```
# usei "count()" para contar as substring A, T, G, C
adenina = s.count("A")
citosina = s.count("C")
guanina = s.count("G")
timina = s.count("T")

print(adenina, citosina, guanina, timina)
```

*Mas qual a utilidade em se contar a quantidade de A, T, G, C?* 

Vamos tomar como exemplo os idiomas que usam o alfabeto latino, como o português, o inglês, o francês e o espanhol. Cada idioma possui um padrão na distribuição do uso de cada letra *(e, se não me engano, essa é uma das maneiras que um editor/tradutor de textos identifica o idioma do texto inserido, e também uma das características que permitem identificar um plágio)*. O genoma de uma dada espécie também possui um padrão nas proporções de A, T, G, C. Portanto, se você sequenciar um genoma desconhecido e confrontá-lo com um banco de dados genômicos, você poderá obter informações sobre a origem desse genoma, como de qual família, gênero, ou até mesmo espécie, ele pertence. Isso porque um dos parâmetros analisados é a quantidade de A, T, G, C. Além disso,  a quantidade desses nucleotídeos também pode ser usada como *feature* para modelos de Machine Learning em genômica.
