#**Contando nucleotídeos do DNA**
___

Nesse exercício, iremos apenas contar a quantidade de nucleotídeos de uma determinada sequência de **DNA** (ácido desoxirribonucleico).

A unidade fundamental do DNA é o **nucleotídeo**, que pode ser dividido em 3 partes:

- um **grupo fosfato**, que confere o caráter ácido à molécula de DNA;

- uma **pentose**, que é um açúcar de 5 carbonos (no caso do DNA, é a desoxirribose: uma pentose que perdeu o oxigênio do carbono 2', *lê-se "carbono 2 linha"*);

- uma **base nitrogenada**, que para o DNA, pode ser **Adenina**,  **Guanina** *(bases púricas: 2 aneis)*, **Citosina** e **Timina** *(bases pirimídicas: 1 anel)*.

Portanto, o DNA é constituído por milhares, milhões, e até bilhões, de nucleotídeos, que podem ser de **quatro tipos**, e estes recebem o nome de acordo com a base nitrogenada que possuírem: **A**denina, **T**imina, **G**uanina ou **C**itosina.

##**Problema:**



http://rosalind.info/problems/dna/

Given: A DNA string s of length at most 1000 nt.

Return: Four integers (separated by spaces) counting the respective number of times that the symbols 'A', 'C', 'G', and 'T' occur in s.

Sample dataset:

```
# string s fornecida como exemplo
s = "AGCTTTTCATTCTGACTGCAACGGGCAATATGTCTCTGTGTGGATTAAAAAAAGAGTGTCTGATAGCAGC"
```

Sample output:

```
# usei "count()" para contar as substring A, T, G, C
adenina = s.count("A")
citosina = s.count("C")
guanina = s.count("G")
timina = s.count("T")

print(adenina, citosina, guanina, timina)
```
