#**Tradução do RNA mensageiro em proteína**
___

Nesse exercício, iremos obter a sequência de aminoácidos da proteína baseada na sequência do RNA mensageiro (mRNA), processo que chamamos de **tradução**.

Da mesma forma que os ácidos nucleicos são formados por nucleotídeos, as proteínas são formadas por **aminoácidos**. Mas como uma sequência de nucleotídeos consegue dar origem a uma proteína? 

Para isso, existe o **Código Genético**, que é o "jeito" que a célula interpreta a sequência de nucleotídeos do mRNA: cada **três nucleotídeos** do mRNA representa **um códon**; cada códon irá codificar um **aminoácido**. Veja na figura abaixo. O códon UUU codifica o aminoácido fenilalamina; o códon CUU, a leucina. 

image.png

Dessa forma, quantos códons são possíveis? Se cada códon é formado por 3 nucleotídeos e existem 4 tipos de nucleotídeos (A, G, C, U):

$4^3 = 64$

temos 64 possibilidades de códons no nosso Código Genético. 

Mas repare que não temos 64 tipos de aminoácidos. Há apenas **20 tipos de aminoácidos**. Isso faz com que nosso Código Genético seja **degenerado** ou **redundante**, ou seja, um aminoácido pode ser codificado por mais de um tipo de códon, porém, um códon traduz apenas um tipo de aminoácido (não é ambíguo). Veja a leucina, que pode ser codificada por seis códons diferentes. Também dizemos que o Código Genético é **universal**, pois esse mesmo código é utilizado por todos os seres vivos (há algumas exceções). 



**>>> Curiosidade: antigamente, a insulina utilizada para o tratamento da diabetes era extraída do pâncreas de porcos e bois. Hoje, já somos capazes de produzir essa insulina artificialmente por meio da tecnologia do DNA Recombinante, na qual o gene humano da insulina foi introduzido no genoma de bactérias e estas, em condições laboratoriais, fabricam a insulina humana. Graças ao Código Genético ser universal, as bactérias são capazes de "ler" o gene da insulina humana inserido em seus genomas e traduzí-lo na proteína insulina.**


Outro detalhe é que existe um códon que irá sinalizar o início da síntese proteica, que chamamos de **start codon** ou **códon de início** (AUG), o qual codifica a **metionina**, e também existem três códons que irão sinalizar o término da tradução, os **stop codons** ou **códons de parada** (UAA, UAG, UGA), que não codificam aminoácido. Logo, temos 61 códons no mRNA que especificam aminoácidos.

image.png


A síntese proteica ocorre dentro do **ribossomo** (figura acima), que irá coordenar todo o processo com o auxílio de outras proteínas. O **RNA transportador** (tRNA) é o responsável por trazer os aminoácidos para a síntese e este possui em uma de suas extremidades o que chamamos de **anticódon**, que é complementar ao códon do mRNA e, dessa forma, fornece o aminoácido correspondente. O aminoácido que chega irá se ligar ao aminoácido da cadeia em formação por meio de uma **ligação peptídica**.

 Uma fita de mRNA será traduzida diversas vezes a fim de formar diversas proteínas, até que seja degradada.


**>>> Curiosidade 2: já que o tRNA é o responsável por trazer o aminoácido para a síntese proteica e visto que existem 61 códons no mRNA que codificam aminoácidos, podemos concluir que existem 61 tipos de tRNA, cada um deles com o anticódon complementar ao códon do mRNA, correto? Porém, a quantidade de tRNA presente nas células é bem menor. O fato é que um tRNA pode reconhecer mais de um tipo de códon. E uma explicação para isso é a hipótese do "emparelhamento incerto" ou "Wobble". Essa hipótese diz que a primeira base do anticódon do tRNA (extremidade 5', lado direito) pode ter um pareamento oscilante com a terceira base do códon do mRNA. Observe a figura abaixo. Se a primeira base do anticódon do tRNA for uma guanina, ela poderá se parear com uma citosina (pareamento normal) ou com uma uracila (pareamento Wobble). Mas repare que o aminoácido codificado continua sendo a leucina. Se você reparar no Código Genético da tabela lá de cima, verá que a maioria dos códons que codificam o mesmo aminoácido diferem apenas na terceira base!**

##**Problema:**

http://rosalind.info/problems/prot/

Given: An RNA string s corresponding to a strand of mRNA (of length at most 10 kbp).

Return: The protein string encoded by s.

Sample Dataset:

```
# string s fornecida pelo exemplo
s = "AUGGCCAUGGCGCCCAGAACUGAGAUCAAUAGUACCCGUAUUAACGGGUGA"
```

Sample Output:

```
# função que recebe o códon e retorna o aminoácido (aa) correspondente
def retorna_aa(codon):

    # dicionário códon:aminoácido
    codon_dict = {"UUU": "F", "CUU": "L", "AUU": "I", "GUU": "V",
                  "UUC": "F", "CUC": "L", "AUC": "I", "GUC": "V",
                  "UUA": "L", "CUA": "L", "AUA": "I", "GUA": "V",
                  "UUG": "L", "CUG": "L", "AUG": "M", "GUG": "V",
                  "UCU": "S", "CCU": "P", "ACU": "T", "GCU": "A",
                  "UCC": "S", "CCC": "P", "ACC": "T", "GCC": "A",
                  "UCA": "S", "CCA": "P", "ACA": "T", "GCA": "A",
                  "UCG": "S", "CCG": "P", "ACG": "T", "GCG": "A",
                  "UAU": "Y", "CAU": "H", "AAU": "N", "GAU": "D",
                  "UAC": "Y", "CAC": "H", "AAC": "N", "GAC": "D",
                  "UAA": "Stop", "CAA": "Q", "AAA": "K", "GAA": "E",
                  "UAG": "Stop", "CAG": "Q", "AAG": "K", "GAG": "E",
                  "UGU": "C", "CGU": "R", "AGU": "S", "GGU": "G",
                  "UGC": "C", "CGC": "R", "AGC": "S", "GGC": "G",
                  "UGA": "Stop", "CGA": "R", "AGA": "R", "GGA": "G",
                  "UGG": "W", "CGG": "R", "AGG": "R", "GGG": "G"}

    aa = codon_dict[codon]

    return aa
```

```
# função principal: recebe a sequência do mRNA e retorna a sequência de aminoácidos
def traducao(s):
    proteina = ""
    tamanho_s = len(s)
    
    # janela de leitura do mRNA, com step = 3
    for pos in range(0, tamanho_s, 3):

        # chamando a função 'retorna_aa(codon)'
        aa = retorna_aa(s[pos:pos+3])

        if aa != "Stop":
            # adiciona o aa à cadeia crescente da proteína
            proteina += aa

        else:
            # término do tradução
            return proteina
```

```
 # chamando a função principal 'traducao(s)'
 traducao(s)
```
